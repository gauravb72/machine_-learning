```python
r=requests.get(url)
soup=BeautifulSoup(r.text,"lxml")
```

```python
for i in range(1,10):
    print("https://www.flipkart.com/search?q=mobiles+under+50000&otracker=search&otracker1=search&marketplace=FLIPKART&as-show=on&as=off&page="+str(i))
```

```python
Product_name1=[]
Prices1=[]
Descriptions1=[]
Reviews1=[]
```

```python
box=soup.find("div",class_="_1YokD2 _3Mn1Gg")
```

```python
Name1=box.find_all('div',class_="_4rR01T")

for i in Name1:
    Names1=i.text
    Product_name1.append(Names1)
    
print(Product_name1)
```

```python
print(len(Product_name1))
```

```python
Price1=box.find_all('div',class_='_30jeq3 _1_WHN1')

for i in Price1:
    Pricess1=i.text
    Prices1.append(Pricess1)
    
print(Prices1)
```

```python
Desc12=box.find_all('ul',class_="_1xgFaf")

for i in Desc12:
    Descr12=i.text
    Descriptions1.append(Descr12)
    
print(Descriptions1)
```

```python
Revie1=box.find_all('div',class_="_3LWZlK")

for i in Revie1:
    Review1=i.text
    Reviews1.append(Review1)
    
print(Reviews1)
```

```python
print(len(Reviews1))
```

```python
print(len(Product_name1))
```

```python
df=pd.DataFrame({'product name':Product_name1,'Prices':Prices1,'Description':Descriptions1,'Reviews':Reviews1})
print(df)
```
